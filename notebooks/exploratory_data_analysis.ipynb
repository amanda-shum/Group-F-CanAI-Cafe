```python
# Convert columns to numeric
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors="coerce")
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors="coerce")

# Only keep rows where Quantity is not missing
df_with_quantity = df[df["Quantity"].notna()].copy()

# Calculate expected total
df_with_quantity["Expected Total Spent"] = (
    df_with_quantity["Quantity"] * df_with_quantity["Price Per Unit"]
)

# Check if Total Spent matches expected total
df_with_quantity["Total Matches"] = np.isclose(
    df_with_quantity["Total Spent"],
    df_with_quantity["Expected Total Spent"],
    atol=0.01
)

# Show rows where Total Spent is incorrect
incorrect_totals = df_with_quantity[df_with_quantity["Total Matches"] == False]

incorrect_totals[
    ["Quantity", "Price Per Unit", "Total Spent", "Expected Total Spent"]
]
```

```python
for col in numeric_columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"{col}: {len(outliers)} possible outliers")
```
