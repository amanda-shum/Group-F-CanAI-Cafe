```python
import pandas as pd
import numpy as np

file_path = "../data/interim/CanAI_Cafe_2023_Sales_Information_Step1.csv"

df = pd.read_csv(file_path)

df.head()
```

```python
df_cleaned = df.copy()

# Convert blank strings / whitespace-only strings to NaN
df_cleaned = df_cleaned.replace(r'^\s*$', np.nan, regex=True)

# Record row count before cleaning
rows_before = len(df_cleaned)

# Drop Payment Method column
df_cleaned = df_cleaned.drop(columns=['Payment Method'], errors='ignore')

df_cleaned['Transaction Date'] = pd.to_datetime(
    df_cleaned['Transaction Date'],
    errors='coerce'
)

# Identify rows with missing Transaction Date or missing Item
rows_to_drop = (
    df_cleaned['Transaction Date'].isna() |
    df_cleaned['Item'].isna()
)

# Optional: keep a copy of dropped rows for audit/debugging
dropped_rows_missing_date_or_item = df_cleaned.loc[rows_to_drop].copy()

# Drop the rows
df_cleaned = df_cleaned.loc[~rows_to_drop].copy()

# Convert Transaction Date to datetime after dropping missing dates
df_cleaned['Transaction Date'] = pd.to_datetime(
    df_cleaned['Transaction Date'],
    errors='coerce'
)

# Summary
rows_after = len(df_cleaned)
rows_dropped = rows_before - rows_after

print(f"Rows before cleaning: {rows_before}")
print(f"Rows after cleaning: {rows_after}")
print(f"Rows dropped: {rows_dropped}")
print(f"Percentage dropped: {rows_dropped / rows_before * 100:.2f}%")

print("\nRemaining missing values:")
print(df_cleaned.isna().sum().sort_values(ascending=False))

df_cleaned.head()
```

```python
import numpy as np
import pandas as pd

df_cleaned = df_cleaned.copy()

# Make sure blank strings are treated as missing
df_cleaned = df_cleaned.replace(r'^\s*$', np.nan, regex=True)

rng = np.random.default_rng(seed=42)

def random_sample_from_series(series, rng):
    """
    Randomly sample one value from the observed distribution of a Series.
    """
    observed = series.dropna()

    if observed.empty:
        return np.nan

    probs = observed.value_counts(normalize=True)

    return rng.choice(
        probs.index,
        p=probs.values
    )


def conditional_random_impute(df, target_col, group_col, rng):
    """
    Impute missing values in target_col by randomly sampling from observed values
    within the same group_col category.

    If the group has no observed values, fall back to global random sampling.
    """
    df = df.copy()

    missing_mask = df[target_col].isna()

    for idx in df[missing_mask].index:
        group_value = df.loc[idx, group_col]

        if pd.notna(group_value):
            group_series = df.loc[
                (df[group_col] == group_value) & (df[target_col].notna()),
                target_col
            ]

            sampled_value = random_sample_from_series(group_series, rng)

            if pd.notna(sampled_value):
                df.loc[idx, target_col] = sampled_value
                continue

        # Fallback: sample from overall distribution
        df.loc[idx, target_col] = random_sample_from_series(df[target_col], rng)

    return df


# Add missingness flags before imputing
df_cleaned['Location_missing_flag'] = df_cleaned['Location'].isna().astype(int)
df_cleaned['Province_missing_flag'] = df_cleaned['Province_cleaned'].isna().astype(int)

# Impute Province based on Location
df_cleaned = conditional_random_impute(
    df=df_cleaned,
    target_col='Province_cleaned',
    group_col='Location',
    rng=rng
)

# Impute Location based on Province
df_cleaned = conditional_random_impute(
    df=df_cleaned,
    target_col='Location',
    group_col='Province_cleaned',
    rng=rng
)

# Check result
print("Remaining missing values:")
print(df_cleaned.isna().sum().sort_values(ascending=False))

print("\nLocation missing flag count:")
print(df_cleaned['Location_missing_flag'].sum())

print("\nProvince missing flag count:")
print(df_cleaned['Province_missing_flag'].sum())

print("\nLocation distribution after imputation:")
print(df_cleaned['Location'].value_counts(normalize=True))

print("\nProvince distribution after imputation:")
print(df_cleaned['Province_cleaned'].value_counts(normalize=True))
```
